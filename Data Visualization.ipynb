# Data Exploration

```python
df = pd.read_csv('data/turkey.csv')
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
plt.figure(figsize=(25,15))
plt.plot(df['Price'])
plt.title('Stock Close price.', fontsize=15)
plt.ylabel('Price in dollars.')
plt.show()
```

```python
df.isnull().sum()
```

Checking for variance in data

```python
features = ['Price', 'Open', 'High', 'Low']
 
plt.subplots(figsize=(20,10))
 
for i, col in enumerate(features):
  plt.subplot(2,3,i+1)
  sb.distplot(df[col])
plt.show()
```

Checking for any outliers

```python
plt.subplots(figsize=(20,10))
for i, col in enumerate(features):
  plt.subplot(2,3,i+1)
  sb.boxplot(df[col])
plt.show()
```
